```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

sns.set_style("dark")
```

```python
superPop = pd.read_csv("E:/Downloads/superPopCluster", sep=" ")
subPop = pd.read_csv("E:/Downloads/subPopCluster", sep=" ")

# CYP2A6:
A6_Vect = pd.read_csv("E:/Downloads/ALL_SUPER_CYP2A6.eigenvec", sep="\t").join(superPop, lsuffix="IID", rsuffix="sample").join(subPop, lsuffix="IID", rsuffix="sample")
A6_Val = pd.read_csv("E:/Downloads/ALL_SUPER_CYP2A6.eigenval", sep=" ")
A6_loadingScores = pd.read_csv("E:/Downloads/ALL_SUPER_CYP2A6.eigenvec.var", sep="\t").drop(["#CHROM", "MAJ", "NONMAJ"], axis='columns').set_index("ID")

# CYP2B6:
B6_Vect = pd.read_csv("E:/Downloads/ALL_SUPER_CYP2B6.eigenvec", sep="\t").join(superPop, lsuffix="IID", rsuffix="sample").join(subPop, lsuffix="IID", rsuffix="sample")
B6_Val = pd.read_csv("E:/Downloads/ALL_SUPER_CYP2B6.eigenval", sep=" ")
B6_Val = pd.read_csv("E:/Downloads/ALL_SUPER_CYP2B6.eigenvec.var", sep="\t")

# UGT2B7:
B7_Vect = pd.read_csv("E:/Downloads/ALL_SUPER_UGT2B7.eigenvec", sep="\t").join(superPop, lsuffix="IID", rsuffix="sample").join(subPop, lsuffix="IID", rsuffix="sample")
B7_Val = pd.read_csv("E:/Downloads/ALL_SUPER_UGT2B7.eigenval", sep=" ")
B7_Val = pd.read_csv("E:/Downloads/ALL_SUPER_UGT2B7.eigenvec.var", sep="\t")
```
